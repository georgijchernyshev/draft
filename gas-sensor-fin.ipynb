```python
z = np.abs(stats.zscore(train_df))
train_df = train_df[(z<3).all(axis=1)]
X = train_df.drop(columns[[0,1,3,4,5,7,8,9,11,14,16,15,18]], axis=1)
Y = train_df.loc[:,[columns[1]]]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
Y = le.fit_transform(Y.values.ravel())
```

```python
from sklearn.decomposition import PCA
svd = PCA(n_components=2)
svd.fit(X)
```

```python
x_trunc_norm_2 = svd.transform(X)

sns.scatterplot(x=x_trunc_norm_2[:, 0], y=x_trunc_norm_2[:, 1], hue=Y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SequentialFeatureSelector as SFS, SelectFromModel
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.preprocessing import StandardScaler, RobustScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, shuffle=True, random_state=42)
```

```python
scaler = StandardScaler()
le = LabelEncoder()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#from sklearn.svm import LinearSVC
#from sklearn.feature_selection import SelectFromModel
#lsvc = LinearSVC(C=0.01, penalty="l1", dual=False, max_iter=10000, verbose=True).fit(X_clean, Y_clean)
#model = SelectFromModel(lsvc, prefit=True)
#X_clean_new = model.transform(X)
```

```python
knn = KNeighborsClassifier(n_jobs=-1)
knn.fit(X_train, y_train)
```

```python
accuracy_score(y_test, knn.predict(X_test)), precision_score(y_test, knn.predict(X_test), average='macro'), recall_score(y_test, knn.predict(X_test), average='macro')
```

```python
rfc = RandomForestClassifier(n_jobs=-1, verbose=True)
rfc.fit(X_train, y_train)
```

```python
accuracy_score(y_test, rfc.predict(X_test)), precision_score(y_test, rfc.predict(X_test), average='macro'), recall_score(y_test, rfc.predict(X_test), average='macro')
```
